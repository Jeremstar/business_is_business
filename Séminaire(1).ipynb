```python
import pandas as pd
from prophet import Prophet
import numpy as np
import math
import pystan
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Jeremstar/business_is_business/main/England_covid.csv')
df.head()
```

```python
#Nettoyage des données
df = df[["date", "new_deaths_60days","new_virus_test","daily_cases"]]
df.head()
```

```python
missing = df.isnull().sum()
print(missing)
print(df.size)

#On décide d'enlever les valeurs manquantes
df = df.dropna(axis=0)
df.head()
```

```python
#On transforme la date en donnée incrémentale 
df = df.sort_values(by='date')
df['j'] = np.arange(df.shape[0])
df["j"] = df['j'] + 1

df = df.sample(300)

df.tail()
```

```python
def find_var_expl_time(j): 
    x_j = [1]
    x_j.append(j/365*3)
    for l in range(2,46): 
        if l >= 2 and l <= 8: 
            x_j.append(math.cos(2*math.pi*(l-2)*j/7))
        elif l >= 9 and l <= 15:
            x_j.append(math.sin(2*math.pi*(l-9)*j/7))
        elif l >= 16 and l <= 30: 
            x_j.append(math.cos(2*math.pi*(l-16)*j/(365.25)))
        elif l >= 31 and l <= 45:
            x_j.append(math.sin(2*math.pi*(l-35)*j/(365.25)))
    return x_j
```

```python
mat_design = np.empty((0,46))
for j in df["j"]:
    row = np.array(find_var_expl_time(j))
    mat_design = np.r_[mat_design,[row]]

mat_design = pd.DataFrame(mat_design)
```

```python
mat_design
```

```python

import warnings
warnings.filterwarnings('ignore')

model_data = {
    'N': df.shape[0],
    'Y': df['new_virus_test'].values,
    'K': mat_design.shape[1],
    'X': mat_design,
    'y' : df["new_virus_test"].astype(int).copy()
}


model_code = """
data {
  int<lower=0> N;   // number of data items
  int<lower=0> K;   // number of predictors
  matrix[N, K] X;   // predictor matrix
  vector[N] Y;      // outcome vector
  int<lower=0> y[N];
}
parameters {
  real alpha;       // intercept
  vector[K] beta;
  real<lower=0> sigma;
}

model {
  for (i in 1:N) {
      y[i] ~ neg_binomial_2(exp(X[i]*beta),sigma); 
      }
}"""

stan_model = pystan.StanModel(model_code=model_code)
fit = stan_model.sampling(data=model_data)

alpha_mean = fit['alpha'].mean()
beta_mean = fit['beta'].mean()
sigma_mean = fit['sigma'].mean()

print(sigma_mean)
```

```python
print("alpha_mean = ",alpha_mean) #-133.3204346579638
print("beta_mean = ",beta_mean) #4.243063695776724
print("beta_mean = ",sigma_mean) #2.0188498034682676e-08
```
